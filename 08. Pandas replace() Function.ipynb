### replace()
by using replace papramets we can replace dataframe value with other value

```python
dataset.replace(to_replace = "Health", value = "Health Insurance")  # replace "Health" with "Health Insurance"
```

or

```python
dataset.replace("heath", "Heath Insurance")       # or simple way to replace first value with second value
```

```python
dataset.replace([1, 2, 3, 4, 5], 'I')      # multiple value replace by with one
```

```python
dataset.replace([1, 2, 3, 4, 5], ['I', 'II', 'III', 'IV', 'V'])      # multiple value replace by using list
```

```python
dataset.replace({"Name" : "Lamtone"}, "Wavy")       # specific value replace using dict in replace method
```

```python
# multiple specific value replace with one value
dataset.replace({"Industry" : "Health", "Inception" : 2009}, "Hello World")
```

```python
# mutliple specific value replace with multiple value
dataset.replace({"Industry" : "Health", "Inception" : 2009}, {"Industry" : "fit", "Inception" : 2008})
```

#### regex
Alternatively, this could be a regular expression used to replace the specific char in string

```python
dataset
```

```python
dataset.replace('L', 'AAAAAAAAA', regex = True)
```

```python
dataset.replace('[A-Za-z]', 0, regex = True)
```

```python
dataset.replace({"Industry" : "IT"}, "Information Technology", regex  = True)
```

#### mehtod
* farward fill (replacepreviuos value with farward None value by individual col)
* farward fill (replace farward value with previuos None value by individual col)

```python
dataset
```

```python
dataset.replace('Health', method = 'ffill')
```

```python
dataset.replace('Financial Services', method = 'bfill')
```

#### limit
In limit we must pass the method or value parameters as well. By using limit parameter we decide how many value we want to fill in col

```python
dataset.replace('Health', 0, limit = 5)
```

#### inplace
it update the dataset with value not create new one

```python
dataset
```

```python
dataset.replace(2009, 2020, inplace = True)
```
