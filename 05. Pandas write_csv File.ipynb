### usecols
by this this papramets we can manually select col that we want to use

### skiprows
by using this parameters we can skip any rows

```python
data_loc = "C:\\Users\\amira\\Data Science\\0. Some  Basic Work\\3. Pandas\\dataset\\Fortune_10_.csv"
dataset = pd.read_csv(data_loc)
dataset
```

```python
dataset = pd.read_csv(data_loc, skiprows = 1)
dataset
```

```python
# or skip row by index
dataset = pd.read_csv(data_loc, skiprows = [0])                 # we skip 0 index of rows
dataset
```

```python
dataset = pd.read_csv(data_loc, skiprows = [0, 10])         # we skip 0 and 10 index of rows
dataset
```

### index_col

```python
dataset = pd.read_csv(data_loc, index_col = 1)          # we make Name as index
dataset
```

### prefix
