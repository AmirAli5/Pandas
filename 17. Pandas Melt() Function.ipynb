### Import Library

### Import Dataset

## Melt Function
Melt function is used to transform or reshape our data

```python
re = pd.melt(dataset)
re
```

### id_vars
by using this parameters we can make any identifier as column

```python
re = pd.melt(dataset, id_vars = ['Class'])
re
```

### value_vars
by using this parameters we can make any identidfiers as a value

```python
re = pd.melt(dataset, value_vars = ['Percantege'])
re
```

### var_name
by using this parameters we can change the variable name like variable >> %

```python
re = pd.melt(dataset, value_vars = ['Percantege'], var_name = "%")
re
```

### value_name
by using this parameters we can change the value name like value >> data

```python
re = pd.melt(dataset, value_vars = ['Percantege'], var_name = "%", value_name = "data")
re
```
