### dropna()
It's basically drop the missing values

```python
data_loc = "C:\\Users\\amira\\Data Science\\0. Some  Basic Work\\3. Pandas\\dataset\\Fortune_102.csv"
dataset = pd.read_csv(data_loc)
dataset
```

```python
dataset.dropna() # it's drop 5, 6, 7, 9 row because of missing  ( by default axis = 1 mean drop missing col)
```

#### dropna axis

```python
dataset.dropna(axis = 1)    # drop missing value (col)
```

#### dropna how

```python
dataset.dropna(how = 'any')    # by default it's any remove missing value
```

```python
dataset.dropna(how = 'all')    # all is remove the missing value if we have all row NaN or Spceifc Col have NaN
```

#### dropna thresh

```python
dataset
```

```python
dataset.dropna(thresh = 8)       # 8 not null value but in row 3, 4, 5, 6, 8 less 8 null value that's they drop
```

#### dropna subset

```python
dataset
```

```python
dataset.dropna( subset = ["Industry"])   # it drove the missing value in Industry Columns
```

### dropna inplace

```python
dataset
```

```python
dataset.dropna( inplace = True)  # It drop the missing value and update the datset not create the new one
```
