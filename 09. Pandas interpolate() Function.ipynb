## Interpolate
Interpolate is basically use to fill the NaN values in Series or DataFrame.<br>
<b>Note:</b> It only fill numeric value not string

### method

#### linear

```python
dataset.interpolate(method='linear')
```

#### nearest

```python
dataset.interpolate(method = 'nearest')
```

#### polynomial

```python
dataset.interpolate(method = 'polynomial', order = 1)
```

```python
dataset.interpolate(method = 'polynomial', order = 2)
```

### Limit
By using this parameters we can fill value by limit in col

```python
dataset.interpolate(mathod = 'linear', limit = 1)
```

```python
dataset.interpolate(mathod = 'linear', limit = 2)
```

```python
dataset.interpolate(mathod = 'linear', limit = 3)
```

```python
dataset.interpolate(mathod = 'linear', limit = 4)
```

### Limit direction
It fill the missing value by direction either is forward, backward or both

```python
dataset.interpolate(limit = 1, limit_direction = 'forward')
```

```python
dataset.interpolate(limit = 1, limit_direction = 'backward')
```

```python
dataset.interpolate(limit = 1, limit_direction = 'both')
```

### Inpalce
It update our dataset

```python
dataset.interpolate(inplace = True)
dataset
```
