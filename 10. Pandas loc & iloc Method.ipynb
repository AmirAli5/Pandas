### Import Dataset

## loc
Access a group of rows and columns by labels

```python
dataset.loc[0]
```

```python
# for multiple we use list of labes
dataset.loc[[0, 2, 4]]
```

```python
# class index 4 value
dataset.loc[4, 'Class']
```

```python
# acess Class col value by using slice
dataset.loc[:, 'Class']
```

```python
# Access all rows of Student ID and Class values
dataset.loc[:, ['Student ID', 'Class']]
```

```python
# Access 2 and 4 rows of Student ID and Class values
dataset.loc[[2, 4], ['Student ID', 'Class']]
```

```python
# access greater than 8 of sleeping hours values
dataset.loc[dataset['Sleeping hrs'] > 8]
```

```python
# access percentage greater than 8 of sleeping hours values
dataset.loc[dataset['Sleeping hrs'] > 8, ['Percantege']]
```

```python
# Access Social Media usage hrs, Mobile Games hrs, Percantege which equal to 11 in class
dataset.loc[dataset['Class'] == 11, ['Social Media usage hrs', 'Mobile Games hrs', 'Percantege']]
```

## iloc
Access a group of rows and columns by Index

```python
dataset
```

```python
# Access Social Media usage hrs col value
dataset.iloc[:, 4]
```

```python
# Access all rows of Student ID and Class values
dataset.iloc[:, 0:2]
```

```python
# Access 2 and 4 rows of Student ID and Class values
dataset.iloc[[2, 4], 0:2]
```

```python
# access all values of class and percentage
dataset.iloc[:, [0, 6]]
```
