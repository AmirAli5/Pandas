### Pivot Table
The level in the pivot table will be stored in Multindex objects on on the index & column of the result DataFrame

#### Index
It's make index of given input 

```python
dataset.pivot_table(index = 'Class')     # by default aggfunc = mean ( and take rest of values mean)
```

#### aggfunc
By default it take mean we can change as well like sum, max, min etc

```python
dataset.pivot_table(index = 'Class', aggfunc = 'sum')
```

```python
dataset.pivot_table(index = 'Class', aggfunc = 'max')
```

#### margin
By passing margin True we can take all value (mean, sum etc)

```python
dataset.pivot_table(index = 'Class', aggfunc = 'sum', margins = True)
```
