## fillna Method

### value
It repalce NaN with Value

```python
dataset.fillna(2)         # NaN value replace with 2
```

```python
dataset.fillna(3)     # replace with 3 
```

```python
# Manually fill Col value use dict
dataset.fillna({"Industry" : "None", "Inception" : 0, "Expenses": 0, "Growth": 0})
```

### method 

#### method = 'ffill'
farward fill (replacepreviuos value with farward None value by individual col)

```python
dataset.fillna(method = 'ffill')
```

#### method = 'bfill'
farward fill (replace farward value with previuos None value by individual col)

```python
dataset.fillna(method = 'bfill')
```

### axis

```python
dataset.fillna(method = 'ffill', axis = 0)       # col
```

```python
dataset.fillna(method = 'ffill', axis = 1)       # row
```

### limit
In  limit we must pass the method or value parameters as well. By using limit parameter we decide how many value we want to fill in col

```python
dataset.fillna(7, limit = 2)    # 2 missing value fill in each col
```

```python
dataset.fillna(7, limit = 3)    # 3 missing value fill in each col
```

```python
dataset.fillna(method = 'ffill', limit = 2)    # with method 
```

### inplace
it update the dataset with value not create new one

```python
dataset.fillna(2, inplace = True)         
dataset
```
